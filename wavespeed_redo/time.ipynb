```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xar
```

```python
ds20 = xar.open_dataset('wavespeed20.nc')
```

```python
ds20_t = ds20.pheight.t

ds20_pheight = ds20.pheight.sel(x=0, y=-0.5)
```

```python
plt.plot(ds20_t, ds20_pheight)
plt.xlabel('t')
plt.ylabel('pheight')
plt.title('Wave height at (x,y) = (0,-0.5)')
plt.show()
```

```python
def turning_points(list):
    dx = np.diff(list)
    return np.where(dx[1:] * dx[:-1] < 0)[0] # added [0] because returns array, therefore extracts

ds_tp = turning_points(ds20_pheight)[0]
ds_tp
```

```python
ds20_t[ds_tp]
```
